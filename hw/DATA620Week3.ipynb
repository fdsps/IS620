```python
import networkx as nx
import matplotlib.pyplot as plt
import graphviz as gv
%matplotlib inline
rhode = nx.read_weighted_edgelist("Rhode.edges", encoding='utf-8')
```

```python
print ("nodes: %d" % rhode.number_of_nodes())
print ("diameter: %d" % nx.diameter(rhode))
print ("radius: %d" % nx.radius(rhode))
print ("eccentricity: %s" % nx.eccentricity(rhode))
print ("center: %s" % nx.center(rhode))
print ("density: %s" % nx.density(rhode))
print ("periphery: %s" % nx.periphery(rhode))
```

Find shortest path between periphery nodes

```python
from networkx import algorithms
path = algorithms.shortest_path(rhode, u'11', u'8')
path
```

```python
edges = zip(path, path[1:])
pos=nx.shell_layout(rhode)
nx.draw_networkx_nodes(rhode, pos=pos)
nx.draw_networkx_labels(rhode, pos=pos)
nx.draw_networkx_edges(rhode, pos=pos)
nx.draw_networkx_edges(rhode, pos=pos, edgelist=edges, edge_color='g', width=3)

```
